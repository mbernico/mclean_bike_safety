```python
import pandas as pd
import seaborn as sns
import os
```

```python
def load_data():
    normal_fn = "../data/derived/normal_il_data.csv" 
    bloomington_fn = "../data/derived/bloomington_il_data.csv" 
    mclean_fn = "../data/derived/mclean_county_data.csv" 

    normal_df = pd.read_csv(normal_fn)
    bloomington_df = pd.read_csv(bloomington_fn)
    mclean_df = pd.read_csv(mclean_fn)
    bike_data = {"normal": normal_df, "bloomington": bloomington_df, "mclean": mclean_df}
    return bike_data
```

```python
def merge_datasets(bike_data):
    df = pd.DataFrame()
    for k,v in bike_data.items():
        v["data_source"] = k
        df = pd.concat([df, v], axis=0)
    return df
```

```python
def clean_date(df):
    df['crash_time'] = pd.to_datetime(df.date + " " + df.time)
    return df
```

```python
bike_data = load_data()
df = merge_datasets(bike_data)
df = clean_date(df)
df = df.drop(['date', 'time'], axis=1)


#print(df.shape)
#print(df.columns)
```
